## 1. Sentiment Analysis

Use the [newsapi](https://newsapi.org/) to pull the latest news articles for Bitcoin and Ethereum and create a DataFrame of sentiment scores for each coin.

Use descriptive statistics to answer the following questions:
1. Which coin had the highest mean positive score?
2. Which coin had the highest negative score?
3. Which coin had the highest positive score?

```python
# Initial imports
import os
import pandas as pd
from datetime import datetime, timedelta
from dotenv import load_dotenv
import nltk as nltk
import alpaca_trade_api as tradeapi
from newsapi.newsapi_client import NewsApiClient
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
%matplotlib inline
```

```python
# Read api key environment variable
load_dotenv("example.env")

# Set Alpaca API key and secret
alpaca_key = os.getenv("ALPACA_KEY")
alpaca_secret = os.getenv("ALPACA_SECRET")

api = tradeapi.REST(alpaca_key, alpaca_secret, api_version='v2')
```

```python
# Create a newsapi client
newsapi = NewsApiClient(api_key=os.environ["NEWS_KEY"])
```

```python
# Fetch the Bitcoin news articles
btc_news_en = newsapi.get_everything(
    q="bitcoin AND Bitcoin",
    language="en",
    page_size = 100,
    sort_by = 'relevancy'
)

# Show the total number of news
btc_news_en["totalResults"]
```

```python
# Fetch the Ethereum news articles
eth_news_en = newsapi.get_everything(
    q="ethereum AND Ethereum",
    language="en",
    page_size = 100,
    sort_by = 'relevancy'
)

# Show the total number of news
eth_news_en["totalResults"]
```

```python
#Create bitcoin sentiment dataframe function
def create_btc_sentiment_df(articles_list):
    sentiments = []

    for article in btc_news_en['articles']:
        try:
            text = article['content']
            date = article['publishedAt'][:10]
            sentiment = analyzer.polarity_scores(text)
            compound = sentiment['compound']
            pos = sentiment['pos']
            neu = sentiment['neu']
            neg = sentiment['neg']
        
            sentiments.append({
                'text':text,
                'date':date,
                'compound':compound,
                'positive':pos,
                'negative':neg,
                'neutral':neu})
        
        except AttributeError:
            pass
    return sentiments
```

```python
# Create the Bitcoin sentiment scores DataFrame
btc_sent_df = create_btc_sentiment_df(btc_news_en)
btc_sent_df = pd.DataFrame(btc_sent_df)
btc_sent_df
```

```python
#Create ethereum sentiment dataframe function
def create_eth_sentiment_df(articles_list):
    sentiments = []

    for article in eth_news_en['articles']:
        try:
            text = article['content']
            date = article['publishedAt'][:10]
            sentiment = analyzer.polarity_scores(text)
            compound = sentiment['compound']
            pos = sentiment['pos']
            neu = sentiment['neu']
            neg = sentiment['neg']
        
            sentiments.append({
                'text':text,
                'date':date,
                'compound':compound,
                'positive':pos,
                'negative':neg,
                'neutral':neu})
        
        except AttributeError:
            pass
    return sentiments
```

```python
# Create the Ethereum sentiment scores DataFrame
eth_sent_df = create_eth_sentiment_df(eth_news_en)
eth_sent_df = pd.DataFrame(eth_sent_df)
eth_sent_df
```

```python
# Describe the Bitcoin Sentiment
btc_sent_df.describe()
```

```python
# Describe the Ethereum Sentiment
eth_sent_df.describe()
```

## 2. Natural Language Processing
---
###   Tokenizer

In this section, you will use NLTK and Python to tokenize the text for each coin. Be sure to:
1. Lowercase each word.
2. Remove Punctuation.
3. Remove Stopwords.

```python
#imports
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from string import punctuation
import re
```

```python
# Instantiate the lemmatizer
lemmatizer = WordNetLemmatizer()

# Create a list of stopwords
nltk.download('stopwords')
sw = set(stopwords.words('english'))
```

```python
# Complete the tokenizer function
def tokenizer(text):
    """Tokenizes text."""
    
    # Remove the punctuation from text
    regex = re.compile("[^a-zA-Z ]")
    re_clean = regex.sub('', text)
   
    # Create a tokenized list of the words
    words = word_tokenize(re_clean)
    
    # Lemmatize words into root words
    lem = [lemmatizer.lemmatize(word) for word in words]
   
    # Convert the words to lowercase and remove stop words
    tokens = [word.lower() for word in lem if word.lower() not in sw]
        
    #return tokens
    return tokens
```

```python
# Create a new tokens column for Bitcoin
btc_sent_df['tokens'] = btc_sent_df.text.apply(tokenizer)
btc_sent_df
```

```python
# Create a new tokens column for Ethereum
eth_sent_df['tokens'] = eth_sent_df.text.apply(tokenizer)
eth_sent_df
```

### NGrams and Frequency Analysis

In this section you will look at the ngrams and word frequency for each coin. 

1. Use NLTK to produce the n-grams for N = 2. 
2. List the top 10 words for each coin. 

```python
#imports
from collections import Counter
from nltk import ngrams
```

```python
# Generate the Bitcoin N-grams where N=2
btc_tokens_list = btc_sent_df['tokens'].sum()
btc_bigrams = Counter(ngrams(btc_tokens_list, n=2))

print(dict(btc_bigrams))
```

```python
# Generate the Ethereum N-grams where N=2
eth_tokens_list = eth_sent_df['tokens'].sum()
eth_bigrams = Counter(ngrams(eth_tokens_list, n=2))

print(dict(eth_bigrams))
```

```python
# Function token_count generates the top 10 words for a given coin
def token_count(tokens, N=3):
    """Returns the top N tokens from the frequency count"""
    return Counter(tokens).most_common(N)
```

```python
# Use token_count to get the top 10 words for Bitcoin
btc_token_count = token_count(btc_bigrams, N=10)
btc_token_count
```

```python
# Use token_count to get the top 10 words for Ethereum
eth_token_count = token_count(eth_bigrams, N=10)
eth_token_count
```

## 3. Named Entity Recognition

In this section, you will build a named entity recognition model for both Bitcoin and Ethereum, then visualize the tags using SpaCy.

```python
#imports
import spacy
from spacy import displacy
```

```python
# Download the language model for SpaCy
!python -m spacy download en_core_web_sm
```

```python
# Load the spaCy model
nlp = spacy.load('en_core_web_sm')
```

---
### Bitcoin NER

```python
# Concatenate all of the Bitcoin text together
btc_joined = btc_sent_df['text'].sum()
btc_joined
```

```python
# Run the NER processor on all of the text
all_btc_text = nlp(btc_joined)

# Add a title to the document
all_btc_text.user_data['title'] = 'Bitcoin NER'
```

```python
# Render the visualization
displacy.render(all_btc_text, style='ent')
```

```python
# List all Entities
for ent in all_btc_text.ents:
    print(f'{ent.text} || {ent.label_}')
```

```python
# Concatenate all of the Ethereum text together
eth_joined = eth_sent_df['text'].sum()
eth_joined
```
